##**Quest√£o 1:** Gere na c√©lula de c√≥digo abaixo as seguintes quantidades de padr√µes (5, 50, 500, 5.000 e 50.000). Avalie numericamente a qualidade da estima√ß√£o de $ \mu $ e $ \Sigma $ obtidas a medida que varia o n√∫mero padr√µes na amosta. Utilize neste experimeto $\mu = \{ 0,0\}$ e $\Sigma \begin{bmatrix}
1.1 & 2.1 \\
2.1 & 5
\end{bmatrix}$.

```python
mu = [0, 0]
Sigma = [[1.1, 2.1], [2.1, 5]]

num_samples = [5, 50, 500, 5000, 50000]
samples = []
for n in num_samples:
    X = np.random.multivariate_normal(mu, Sigma, n)
    samples.append(X)

estimated_mu = []
estimated_Sigma = []
for X in samples:
    m_x = np.mean(X, axis=0)
    S_x = np.cov(X.T)
    estimated_mu.append(m_x)
    estimated_Sigma.append(S_x)

for i, n in enumerate(num_samples):
    mu_error = np.mean((estimated_mu[i] - mu)**2)
    Sigma_error = np.mean((estimated_Sigma[i] - Sigma)**2)
    print(f'Erro para {n} amostras:')
    print(f'Erro em mu: {mu_error}')
    print(f'Erro em Sigma: {Sigma_error}\n')
```

#**Quest√£o 2**: Estime os valores de $ \mu $ e $ \Sigma $ para as amostras X e Y. Em seguida, utilizando probabilidades a priori iguais para ambas as classes, insira no gr√°fico acima as fronteiras de decis√£o obtidas para o classificador Bayesiano para os valores estimados e para os valores verdadeiros.   

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Dados fornecidos
mu_x = [0, 0]
mu_y = [4, 3]
Sigma = [[1.1, 2.1], [2.1, 5]]
X = np.random.multivariate_normal(mu_x, Sigma, 200)
Y = np.random.multivariate_normal(mu_y, Sigma, 200)

# Estimar mu e Sigma para X e Y
m_x = np.mean(X, axis=0)
S_x = np.cov(X.T)
m_y = np.mean(Y, axis=0)
S_y = np.cov(Y.T)

# Fun√ß√£o para calcular a fronteira de decis√£o Bayesiana
def decision_boundary(x, mu1, mu2, sigma):
    """Calcula a fun√ß√£o discriminante para a fronteira de decis√£o."""
    inv_sigma = np.linalg.inv(sigma)
    part1 = (x - mu1).reshape(1, -1) @ inv_sigma @ (x - mu1).reshape(-1, 1)
    part2 = (x - mu2).reshape(1, -1) @ inv_sigma @ (x - mu2).reshape(-1, 1)
    return part1 - part2

# Criar a grade de pontos para o gr√°fico
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
grid_points = np.c_[xx.ravel(), yy.ravel()]

# Calcular a fronteira de decis√£o com valores verdadeiros
Z_true = np.array([decision_boundary(p, mu_x, mu_y, Sigma) for p in grid_points])
Z_true = Z_true.reshape(xx.shape)

# Calcular a fronteira de decis√£o com valores estimados
Z_est = np.array([decision_boundary(p, m_x, m_y, S_x) for p in grid_points])
Z_est = Z_est.reshape(xx.shape)


# Plotar as amostras e as fronteiras de decis√£o
plt.figure(figsize=(10, 10))
plt.plot(X[:, 0], X[:, 1], 'xk', label='Amostra X')
plt.plot(Y[:, 0], Y[:, 1], 'xr', label='Amostra Y')
plt.contour(xx, yy, Z_true, [0], colors='b', label='Fronteira Verdadeira')
plt.contour(xx, yy, Z_est, [0], colors='g', linestyles='--', label='Fronteira Estimada')
plt.axis('equal')
plt.legend()
plt.title('Fronteiras de Decis√£o Bayesianas')
plt.show()

# Imprimir os resultados
print("Valores estimados para X:")
print("mu_x:", m_x)
print("Sigma_x:\n", S_x)

print("\nValores estimados para Y:")
print("mu_y:", m_y)
print("Sigma_y:\n", S_y)
```

#**Quest√£o 3**: Trace em um gr√°fico as fronteiras de decis√£o para o problema de duas classes anterior usando os valores verdadeiros de $\mu$ e $\Sigma$ variando os valores das probabilidades a priori, trace as fronteiras de decis√£o para a probabilidade a priori $P(X)$ igual a $\{0.2, 0.4, 0.6 \, \textrm{e} \, 0.8\}$.
PS.: N√£o esque√ßa de variar a probabilidade a priori de um objeto pretencer √† classe Y de forma coerente.

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados fornecidos (arredondar para 3 casas decimais)
mu_x = [-0.122, -0.196]
mu_y = [3.938, 2.891]
Sigma = [[1.1, 2.1], [2.1, 5]]

# Probabilidades a priori para X
prior_probs_X = [0.2, 0.4, 0.6, 0.8]

# Criar a grade de pontos para o gr√°fico
x_min, x_max = -4, 8
y_min, y_max = -4, 8
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
grid_points = np.c_[xx.ravel(), yy.ravel()]

# Fun√ß√£o para calcular a fun√ß√£o discriminante (sem scipy.stats)
def discriminant_func(x, mu, Sigma, prior_prob):
    inv_Sigma = np.linalg.inv(Sigma)
    centered_x = x - mu
    discriminant_value = (-0.5 * (centered_x.T @ inv_Sigma @ centered_x) +
                          np.log(prior_prob) - 0.5 * np.log(np.linalg.det(Sigma)))

    return discriminant_value


# Plotar as fronteiras de decis√£o para cada probabilidade a priori
plt.figure(figsize=(10, 10))

for prior_prob_X in prior_probs_X:
    # Probabilidade a priori para Y
    prior_prob_Y = 1 - prior_prob_X

    # Calcular os valores da fun√ß√£o discriminante para os pontos da grade
    Z = np.array([discriminant_func(p, mu_x, Sigma, prior_prob_X) -
                  discriminant_func(p, mu_y, Sigma, prior_prob_Y)
                  for p in grid_points])

    Z = Z.reshape(xx.shape)

    # Plotar a fronteira de decis√£o
    plt.contour(xx, yy, Z, levels=[0],
                colors=['r', 'g', 'b', 'm'][prior_probs_X.index(prior_prob_X)],
                label=f'P(X) = {prior_prob_X}')

# Plotar os pontos das classes X e Y
X = np.random.multivariate_normal(mu_x, Sigma, 200)
Y = np.random.multivariate_normal(mu_y, Sigma, 200)
plt.plot(X[:, 0], X[:, 1], 'xk', label='Amostra X')
plt.plot(Y[:, 0], Y[:, 1], 'xr', label='Amostra Y')

plt.axis('equal')
plt.title('Fronteiras de Decis√£o Bayesianas com Diferentes Probabilidades a Priori')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

**Fronteiras de Decis√£o**:
- Vermelha: P(x) = 0.2
- Verde: P(x) = 0.4
- Azul: P(x) = 0.6
- Magenta: P(x) = 0.8


##**Quest√£o 4**: Projete as amostras X e Y na primeira PCA. Obtenha a fronteira de decis√£o no espa√ßo da PCA para o classificador Bayesiano utilizando probabilidades a priori iquais a 0.5. Apresente seus resultados graficamente, utilizando a c√©lula de c√≥digo abaixo.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.stats import norm

# Dados fornecidos (arredondar para 3 casas decimais)
mu_x = [-0.122, -0.196]
mu_y = [3.938, 2.891]
Sigma = [[1.1, 2.1], [2.1, 5]]
X = np.random.multivariate_normal(mu_x, Sigma, 200)
Y = np.random.multivariate_normal(mu_y, Sigma, 200)

# Projetar amostras na primeira componente pricipal (PCA)
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)
Y_pca = pca.transform(Y)

# Calcular a m√©dia e o desvio padr√£o das proje√ß√µes
X_mean = np.mean(X_pca)
X_std = np.std(X_pca)
Y_mean = np.mean(Y_pca)
Y_std = np.std(Y_pca)

# Obter a fronteira de decis√£o
x_min = min(X_pca.min(), Y_pca.min()) - 1
x_max = max(X_pca.max(), Y_pca.max()) + 1
x_range = np.linspace(x_min, x_max, 100)

# Probabilidades a priori (iguais a 0.5)
prior_prob_X = 0.5
prior_prob_Y = 0.5

# Fun√ß√£o discriminante (Bayesiana)
discriminant_func = lambda x: (norm.pdf(x, X_mean, X_std) * prior_prob_X -
                                norm.pdf(x, Y_mean, Y_std) * prior_prob_Y)
decision_boundary = x_range[np.argmin(np.abs(discriminant_func(x_range)))]

# Plotar os resultados
plt.figure(figsize=(8, 6))
plt.hist(X_pca, bins=20, alpha=0.5, label='Classe X', color='black')
plt.hist(Y_pca, bins=20, alpha=0.5, label='Classe Y', color='red')
plt.axvline(x=decision_boundary, color='blue', linestyle='--', label='Fronteira de Decis√£o')
plt.xlabel('Primeira Componente Principal (PCA)')
plt.ylabel('Frequ√™ncia')
plt.title('Proje√ß√£o das Amostras e Fronteira de Decis√£o na PCA')
plt.legend()
plt.show()
```

#**Quest√£o 5**: Fa√ßa o mesmo que fora solicitado na quest√£o 4, mas ao inv√©s de PCA use o LDA.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from scipy.stats import norm

# Dados fornecidos (arredondar para 3 casas decimais)
mu_x = [-0.122, -0.196]
mu_y = [3.938, 2.891]
Sigma = [[1.1, 2.1], [2.1, 5]]
X = np.random.multivariate_normal(mu_x, Sigma, 200)
Y = np.random.multivariate_normal(mu_y, Sigma, 200)

# Criar e treinar o modelo LDA
lda = LinearDiscriminantAnalysis(n_components=1, priors=[0.5, 0.5])
lda.fit(np.concatenate((X, Y)), np.concatenate((np.zeros(len(X)), np.ones(len(Y)))))

# Projetar as amostras no LDA
X_lda = lda.transform(X)
Y_lda = lda.transform(Y)

# Calcular a m√©dia e o desvio padr√£o das proje√ß√µes
X_mean = np.mean(X_lda)
X_std = np.std(X_lda)
Y_mean = np.mean(Y_lda)
Y_std = np.std(Y_lda)

# Obter a fronteira de decis√£o
x_min = min(X_lda.min(), Y_lda.min()) - 1
x_max = max(X_lda.max(), Y_lda.max()) + 1
x_range = np.linspace(x_min, x_max, 100)

# Probabilidades a priori (iguais a 0.5)
prior_prob_X = 0.5
prior_prob_Y = 0.5

# Fun√ß√£o discriminante (Bayesiana)
discriminant_func = lambda x: (norm.pdf(x, X_mean, X_std) * prior_prob_X -
                                norm.pdf(x, Y_mean, Y_std) * prior_prob_Y)
decision_boundary = x_range[np.argmin(np.abs(discriminant_func(x_range)))]

# Plotar os resultados
plt.figure(figsize=(8, 6))
plt.hist(X_lda, bins=20, alpha=0.5, label='Classe X', color='black')
plt.hist(Y_lda, bins=20, alpha=0.5, label='Classe Y', color='red')
plt.axvline(x=decision_boundary, color='blue', linestyle='--', label='Fronteira de Decis√£o')
plt.xlabel('An√°lise de Discriminante Linear (LDA)')
plt.ylabel('Frequ√™ncia')
plt.title('Proje√ß√£o das Amostras e Fronteira de Decis√£o no LDA')
plt.legend()
plt.show()
```
